```python
#resource: https://gist.github.com/bikz05/6fd21c812ef6ebac66e1

def rgb2lab(rgb):
    #Conversion Matrix
    matrix = [[0.412453, 0.357580, 0.180423],
             [0.212671, 0.715160, 0.072169],
             [0.019334, 0.119193, 0.950227]];
    
    rgb = rgb/255;
    cie = np.dot(matrix, rgb)
    
    cie[0] = cie[0]/0.950456
    cie[2] = cie[2]/1.088754 
    
    # Calculate the L
    L = (116 * np.power(cie[1], 1/3.0) - 16.0) if cie[1] > 0.008856 else (903.3 * cie[1])
    
    cie[0] = np.power(cie[0], 1/3.0) if cie[0] > 0.008856 else 7.787 * cie[0] + (16.0/116.0)
    cie[1] = np.power(cie[1], 1/3.0) if cie[1] > 0.008856 else 7.787 * cie[1] + (16.0/116.0)
    cie[2] = np.power(cie[2], 1/3.0) if cie[2] > 0.008856 else 7.787 * cie[2] + (16.0/116.0)

    # Calculate the a 
    a = 500*(cie[0] - cie[1]) + 128.0

    # Calculate the b
    b = 200*(cie[1] - cie[2]) + 128.0

    #  Values lie between -128 < b <= 127, -128 < a <= 127, 0 <= L <= 100 
    Lab = [b , a, L]; 

    # OpenCV Format
    L = L * 2.55
    a = a + 128
    b = b + 128
    Lab_OpenCV = [b, a, L]
    
    return Lab
```
